```python
class User:
  """Represents a user in the social media system."""
  def __init__(self, username):
    self.username = username
    self._posts = []

  def create_post(self, content):
    """
    Creates a new Post object and stores it.
    Returns the created Post so it can be used immediately.
    TODO: Complete this method
    - Create a Post object with self as author
    - Store it in self._posts list
    - Return the created Post
    """
    post = Post(self, content)
    self._posts.append(post)
    return post

  def get_posts(self):
    """Returns all posts created by this user."""
    return self._posts

class Post:
  """Represents a text post in the feed."""
  def __init__(self, author, content):
    """
    author: User object who created this post
    content: string containing the post text
    TODO: Initialize like counter
    - Create an instance variable to track likes
    - Initialize it to 0
    - Must persist across multiple like() calls
    """
    self.author = author
    self.content = content
    self.likes = 0 #LR: I added the like counter, like(self) must increment this

  def like(self):
    """
    Increments the like count.
    Must update internal state properly.
    TODO: Implement this method
    - Increment the like counter by 1
    """
    self.likes += 1 #WR: I increment the like count by 1 here

  def like_count(self):
    """Returns the current number of likes for this post."""
    # TODO: Return the like counter value
    return self.likes #WR: I return the like count here

  def display(self):
    """Displays the post in a readable format."""
    print(f"{self.author.username}: {self.content} ❤ {self.like_count()}")

class Reel:
  """Represents a video reel in the feed."""
  def __init__(self, author, title, duration):
    self.author = author
    self.title = title
    self.duration = duration

  def display(self):
    """Displays the reel."""
    print(f"﻿ Reel: {self.title} ({self.duration}s)")

class Ad:
  """Represents a sponsored advertisement in the feed."""
  def __init__(self, brand, message):
    self.brand = brand
    self.message = message

  def display(self):
    """Displays the advertisement."""
    print(f"﻿ {self.brand}: {self.message}")

def build_feed(items):
  """
  Generator function that yields content items for display.
  Implements lazy loading - items are not all loaded at once.
  Supports repeated iteration over the same items.
  TODO: Implement as a generator function
  - Iterate through the items list
  - Yield each item one at a time
  - This allows the feed to work with any number of items efficiently
  """
  for item in items: #WR: This should be the generator that the TO DO asks for
    yield item

class Feed:
  """
  Container for feed items that can be iterated.
  Implements the iterator protocol to support for-loops.
  """
  def __init__(self, items):
    self.items = items

  def __iter__(self):
    """
    Makes Feed work in a for-loop by implementing the iterator
    protocol.
    TODO: Implement the iterator protocol
    - Create and return an iterator object
    - This allows: for item in feed: ...
    - The iterator should go through self.items
    Hint: You might create an index-based iterator or use
    iter(self.items)
    """
    return iter(self.items) #WR: I return the iterator here

class Analytics:
  """
  Tracks interactions (like, view, etc.) without using global state.
  Each Analytics instance tracks independently.
  No global counters - all state is stored in the instance.
  """
  def __init__(self):
    """
    Initialize the tracking mechanism.
    TODO: Set up tracking
    - Create a way to track counts for different actions
    - Could use a dictionary: {"like": 0, "view": 0, ...}
    - Or use instance variables
    - No global variables allowed
    """
    self.tracking = {}
  def record(self, action):
    """
    Records an action and returns the new count.
    TODO: Implement action recording
    - Increment the count for this action type
    - Return the new count value
    - Must work for any action string
    """
    if action in self.tracking:
      self.tracking[action] += 1
    else:
      self.tracking[action] = 1
    return self.tracking[action]

def log_action(action_name):
  """
  Decorator that logs when an action occurs.
  Must print [LOG] action_name before executing the function.
  Then executes the original function.
  TODO: Implement this decorator
  - Takes action_name as parameter
  - Returns a decorator function
  - Decorator wraps the target function
  - Prints: [LOG] action_name
  - Calls the wrapped function
  - Returns the result
  Example:
  @log_action("like")
  def like_post(post):
  post.like()
  When called, should print: [LOG] like
  """
  def decorator(func):
    def wrapper(*args, **kwargs):
      print(f"[LOG] {action_name}")
      return func(*args, **kwargs)
    return wrapper
  return decorator

# Example usage (these will be decorated):
@log_action("create_post")
def create_post(user, content):
  """Create a post and log the action and then display the post."""
  post = user.create_post(content)
  post.display()
  return post

@log_action("like")
def like_post(post):
  """Like a post, log the action, and then display the updated post."""
  post.like()
  post.display()
```

```python
def main():
  # Create users
  alice = User("alice")
  bob = User("bob")
  charlie = User("charlie")

  # Create some posts"
  post1 = create_post(alice, "Hello World!")
  post2 = create_post(bob, "Just vibing")
  post3 = create_post(charlie, "Python is fun")

  # Like some posts
  like_post(post1)
  like_post(post1)
  like_post(post2)

  # Create other content types
  reel = Reel(alice, "Summer Fun", 30)
  ad = Ad("iPhone", "New iPhone 15 available now!")

  # Build feed with all items
  all_items = [post1, post2, post3, reel, ad]

  # Display using Feed class iterator
  print("\n--- Feed Display (using Feed iterator) ---")
  feed = Feed(all_items)
  for item in feed:
    item.display()

  # Display using build_feed generator
  print("\n--- Feed Display (using build_feed generator) ---")
  for item in build_feed(all_items):
    item.display()

  # Test Analytics
  print("\n--- Analytics Test ---")
  analytics = Analytics()
  print(f"First like: {analytics.record('like')}") # Should print: 1
  print(f"Second like: {analytics.record('like')}") # Should print: 2
  print(f"First view: {analytics.record('view')}") # Should print: 1
  print(f"Third like: {analytics.record('like')}") # Should print: 3

if __name__ == "__main__":
  main()
```
